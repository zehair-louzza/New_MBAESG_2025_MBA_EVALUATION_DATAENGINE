# 🔍 02 — Analyse Exploratoire (EDA)
Distributions, corrélations et profil du dataset.

```python
from snowflake.snowpark.context import get_active_session
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns

session = get_active_session()
session.sql("USE DATABASE HOUSE_PRICE_DB").collect()

df = session.table("BRONZE.HOUSE_PRICE_STRUCTURED").to_pandas()
print(f"Shape : {df.shape}")
df.head()
```

```python
# Types et valeurs manquantes
print(df.dtypes)
print("\nNull counts :")
print(df.isnull().sum())
```

```python
# Distribution du PRICE
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(df['PRICE'], bins=30, color='#01696f', edgecolor='white')
axes[0].set_title("Distribution PRICE"); axes[0].set_xlabel("Prix")
df['PRICE'].plot(kind='box', ax=axes[1], color='#01696f')
axes[1].set_title("Boxplot PRICE")
plt.tight_layout(); plt.show()
```

```python
# Corrélations numériques
df_num = df[['PRICE','AREA','BEDROOMS','BATHROOMS','STORIES','PARKING']].copy()
corr = df_num.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title("Matrice de corrélation")
plt.tight_layout(); plt.show()
```

```python
# Prix selon variables catégorielles
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for ax, col in zip(axes, ['AIRCONDITIONING','FURNISHINGSTATUS','PREFAREA']):
    df.boxplot(column='PRICE', by=col, ax=ax)
    ax.set_title(f'PRICE par {col}')
plt.tight_layout(); plt.show()
```
